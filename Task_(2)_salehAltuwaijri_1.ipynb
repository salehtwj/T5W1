# Tasks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset.csv')
df
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
corrdf = df[['TrafficIndexLive','JamsCount', 'JamsDelay','JamsLength' ,'TrafficIndexWeekAgo' ,'TravelTimeHistoric','TravelTimeLive']]
```

```python
corrdf.corr()
```

### Identify and analyze patterns in the data.

```python
temp =(
    corrdf.loc[:,['TrafficIndexLive','JamsCount', 'JamsDelay','JamsLength' ,'TrafficIndexWeekAgo' ,'TravelTimeHistoric','TravelTimeLive']]
).corr()
sns.heatmap(temp, annot=True)
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df['JamsCount'].value_counts().sort_index().plot(kind='bar')
plt.title('Traffic count ')
plt.show()
```

```python
df['JamsCount'].value_counts().sort_index().plot(kind='bar')
plt.title('Traffic Jams Count')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], bins=30, kde=True)
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.title('Distribution of Traffic Index Live')
plt.show()
```

```python
city_traffic = df.groupby('City')['TrafficIndexLive'].sum().sort_values()
```

### we will grup the traffficusing the city to we can see the city with highst traffic index live

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
plt.figure(figsize=(12, 8))
city_traffic.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Total Traffic Index')
plt.title('Total Traffic Index Live by City')
plt.xticks(rotation=90)
plt.show()
```
